```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

# Datos de calificaciones de los estudiantes
matematicas = rng.integers(50, 100, 20)
ciencias = rng.integers(40, 95, 20)
literatura = rng.integers(60, 100, 20)

# Datos de notas promedio
promedio_matematicas = np.mean(matematicas)
promedio_ciencias = np.mean(ciencias)
promedio_literatura = np.mean(literatura)

# Datos de errores para el gráfico de barras de error
errores_matematicas = rng.uniform(2, 5, 2)
errores_matematicas = [min(errores_matematicas), max(errores_matematicas)]

errores_ciencias = rng.uniform(1, 4, 2)
errores_ciencias = [min(errores_ciencias), max(errores_ciencias)]

errores_literatura = rng.uniform(3, 6, 2)
errores_literatura = [min(errores_literatura), max(errores_literatura)]
```

## Gráfico de dispersión para Matemáticas vs Ciencias

```python
plt.figure(figsize=(8, 6))
plt.scatter(matematicas, ciencias, color='blue', alpha=0.7)  # El marcador por defecto es un punto
plt.title("Relación entre las Calificaciones de Matemáticas y Ciencias")
plt.xlabel("Calificaciones de Matemáticas")
plt.ylabel("Calificaciones de Ciencias")
plt.show()
```

## Crear el gráfico de dispersión con barras de error ajustadas


```python
materias = ["Matemáticas", "Ciencias", "Literatura"]
bar_positions = np.arange(len(materias))

# Crear una lista de promedios
promedios = [promedio_matematicas, promedio_ciencias, promedio_literatura]

# Ajustar los errores para ser simétricos en relación con los promedios
errores = np.array([errores_matematicas, errores_ciencias, errores_literatura]).T

#linea_con_error, = plt.errorbar(x=bar_positions, y=promedios, yerr=errores, fmt='o', capsize=5, label='Promedio con Error')

plt.figure(figsize=(8, 6))
plt.errorbar(x=bar_positions, y=promedios, yerr=errores, fmt='o', capsize=5, label='Promedio')

# Personalización del gráfico
plt.xticks(bar_positions, materias)
plt.title("Calificaciones Promedio con Barras de Error en Diferentes Materias")
plt.xlabel("Materias")
plt.ylabel("Calificaciones Promedio")
plt.ylim(65, 85)  # Establecer límites para el eje y
#plt.grid(True, axis='y')
plt.legend(loc='upper left')  # Ajusta la posición según necesites


plt.show()
```

## Crear un histograma para la distribución de las calificaciones de Matemáticas

```python
plt.figure(figsize=(8, 6))
plt.hist(matematicas, bins=10)

# Personalización del histograma
plt.title("Distribución de las calificaciones de Matemáticas")
plt.xlabel("Calificaciones de Matemáticas")
plt.ylabel("Frecuencia")
#plt.grid(axis='y', alpha=0.75)

plt.show()
```
